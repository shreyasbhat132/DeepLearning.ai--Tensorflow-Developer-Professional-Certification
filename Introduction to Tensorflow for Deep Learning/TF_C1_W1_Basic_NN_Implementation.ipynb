# Simple Housing Price Prediction
Consider the problem where the base price of the house is 50k, and for every additional room the price increases by another 50k. Build a NN to predict the price.


```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
```

Define samples:

```python
x_train = [1, 2, 3, 4, 5, 6]
y_train = [1, 1.5, 2, 2.5, 3, 3.5]
x_test = [10, 7, 8, 9, 11, 13]
y_test = [5.5, 4, 4.5, 5, 6, 7]
```

Fit the model and compile:

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
history = model.fit(x_train, y_train, epochs=300)
```

Predicting price with a test set

```python
model.predict(x_test)
```
